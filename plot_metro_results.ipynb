```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Tool used to plot the graphs from the result files
raw_experiment_data = pd.read_csv('results/metro/lin reg extra1621935722.702808.csv', header=None)
print(raw_experiment_data)
npdata = raw_experiment_data.to_numpy()
print(npdata)
line_names = {
    1: 'temperature',
    2: 'rain_1h',
    3: 'snow_1h',
    4: 'temperature + rain_1h',
    5: 'temperature + snow_1h',
    6: 'rain_1h + snow_1h',
    7: 'temperature + rain_1h + snow_1h'
}
input_variances = np.arange(0,0.4, 0.01)

for i in range(0, npdata.shape[0]):
    print(i)
    plt.plot(input_variances, npdata[i], label=line_names[i+1])
# plt.title('(b)',y=-0.7)
plt.xlabel('Variance in noise',fontsize=20)
plt.ylabel('Output variance',fontsize=20)
plt.tick_params(axis='x', labelsize=13)
plt.tick_params(axis='y', labelsize=13)
lgd = plt.legend(title="Attributes with noise", loc='upper center', bbox_to_anchor=(0.5,-0.2),ncol=1, prop={'size': 14})
plt.savefig('results/metro/lin reg zoom out.png', bbox_inches='tight')

plt.show()

```
