```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Tool used to plot the graphs from the result files
raw_experiment_data = pd.read_csv('results_quantitative/community/FINAL COMM1624710019.425393.csv', header=None)
print(raw_experiment_data)
npdata = raw_experiment_data.to_numpy()
print(npdata)
line_names = {
    1: 'percentage of people under poverty line',
    2: 'median income',
    3: 'percentage of black people',
    4: 'percentage of people under poverty line + median income',
    5: 'percentage of people under poverty line + percentage of black people',
    6: 'median income + percentage of black people',
    7: 'percentage of people under poverty line + median income + percentage of black people'
}
input_variances = np.arange(0,0.01, 0.001)

for i in range(0, npdata.shape[0]):
    print(i)
    plt.plot(input_variances, npdata[i], label=line_names[i+1])
plt.title('Results for changing the percentage of corrupted entries in the attributes with noise \n with fixed variance of 0.001 to investigate the large increase in output variance \n for regression trees (Community data set). Percentages varying from 0% to 10% \n with steps of 1%')
plt.xlabel('Variance in noise',fontsize=20)
plt.ylabel('Output variance',fontsize=20)
plt.tick_params(axis='x', labelsize=13)
plt.tick_params(axis='y', labelsize=13)
lgd = plt.legend(title="Attributes with noise", loc='upper center', bbox_to_anchor=(0.5,-0.2),ncol=1, prop={'size': 10})
plt.savefig('results_quantitative/community/FINAL COMM.png',  bbox_inches='tight')
plt.show()

```
